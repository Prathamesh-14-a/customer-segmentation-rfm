## Data Cleaning

* Checking for a Nulls in Customer_ID

```python
# Counting all CustomerID including nulls 
df['CustomerID'].size
```

```python
#Retriving Data With Nulls CustomerID's
df[df['CustomerID'].isnull()]
```

```python
#Counting all Null CustomerID's
df.CustomerID.isnull().sum()
```

```python
# Removing Rows having Null customerID
df = df[df['CustomerID'].notna()].copy()
```

```python
print(f"Number of CustomerID's after removing Nulls: {df.shape[0]}")
```

* Removing cancelled transactions

In Dataset's InvoiceNo field, cancelled invoice numbers are mentioned by placing 'C' before invoice number so, We will focus to remove invoice numbers which starts with 'C'.

```python
#Cheking for Cancelled invoice numbers
df[df['InvoiceNo'].str.startswith("C" , na = False)]
```

```python
#Taking non canceled InvoiceNo into consideration
df_filtered = df[~df['InvoiceNo'].str.startswith("C" , na= False)].copy()
```

```python
df_filtered
```

```python
df_filtered.shape
```

* Removing Rows having Quantity ≤ 0 & UnitPrice ≤ 0

```python
#Retriving columns having Quantities less than or equal to zero
df_filtered[df_filtered['Quantity'] <= 0]
```

There are 0 rows having quantity less than 0 so no need to perform removal operation.

```python
#retriving columns having UnitPrice less than or equal to zero 

df_filtered[df_filtered['UnitPrice'] <= 0].head()
```

```python
df_filtered[df_filtered['UnitPrice'] <= 0].shape[0]
```

```python
#Taking Rows which have UnitPrice and Quantity greater than 0
df_filtered = df_filtered[df_filtered['UnitPrice'] > 0].copy()
```

```python
# No of rows after removing 
df_filtered.shape[0]
```

* Removing Duplicate rows

```python
df_filtered.duplicated().sum()
```

```python
df_cleaned = df_filtered.drop_duplicates().copy()
```

```python
print(f"Total rows after removing duplicates: {df_cleaned.shape[0]}")
```

* Removing extra and trailing spaces from columns

```python
df_cleaned['Country'] = df_cleaned['Country'].str.strip()
df_cleaned['Description'] = df_cleaned['Description'].str.strip()
```

Cleaned categorical data by stripping whitespace from country names and product description to prevent duplicate categories and improve data integrity.

# Creating New Feature

TotalAmount = Quantity × UnitPrice


```python
df_cleaned['TotalAmount'] = df_cleaned['Quantity'] * df_cleaned['UnitPrice']
```

 created a new feature called Total Amount by aggregating customer transaction values. This feature represents the total monetary value spent by each customer and is used as the Monetary component in the RFM analysis to measure customer value.

* Check Monetary outliers

```python
sns.histplot(df_cleaned["TotalAmount"], bins=50)
plt.xlim(0, df_cleaned["TotalAmount"].quantile(0.99))
```

Most of the customers around 99% are tiny buyers in range of 0 to 200

```python
Q1 = df_cleaned["TotalAmount"].quantile(0.25)
Q3 = df_cleaned["TotalAmount"].quantile(0.75)
IQR = Q3 - Q1

df_cleaned[df_cleaned["TotalAmount"] > Q3 + 1.5 * IQR]
```

```python
sns.boxplot(x=df_cleaned["TotalAmount"])
```

* 99% of transactions are tiny, A few transactions are insanely large (50k, 75k, 170k).These are outliers, not normal customers

If we keep them:

Monetary gets dominated by 2–3 customers

Champions = whoever had one huge order

RFM becomes meaningless

```python
upper = df_cleaned['TotalAmount'].quantile(0.995)
df_cleaned['TotalAmount'] = df_cleaned['TotalAmount'].clip(upper=upper)
```

“Extreme transaction values were analyzed during EDA to understand distribution skewness.
For RFM scoring, raw Monetary values were retained to preserve true customer value.”

```python
sns.boxplot(x=df_cleaned["TotalAmount"])
```

* after capping extreme transaction values , we can see natural right-skewed purchasing behavior

# EDA Summary & Observations

* The dataset contains a large number of repeat customers, with purchasing behavior heavily skewed toward small transaction values and a small number of high-value orders.

* After cleaning, extreme transaction values were capped to prevent bulk purchases from distorting monetary-based analysis.

* The majority of revenue is generated by a small subset of customers, confirming the need for customer-level aggregation rather than transaction-level analysis.

* Country-wise revenue distribution shows clear dominance by a few countries, which will be considered during segmentation analysis.

* The cleaned data is now consistent, free of invalid transactions, and suitable for computing Recency, Frequency, and Monetary (RFM) metrics.
